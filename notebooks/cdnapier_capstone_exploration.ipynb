```python
def get_overall_score(cost, value):
    if ('Less' in cost) & ('Better' in value):
        return 9
    elif ('No ' in cost) & ('Better' in value):
        return 8
    elif ('Greater' in cost) & ('Better' in value):
        return 7
    elif ('Less' in cost) & ('Average' in value):
        return 6
    elif ('No ' in cost) & ('Average' in value):
        return 5
    elif ('Greater' in cost) & ('Average' in value):
        return 4
    elif ('Less' in cost) & ('Worse' in value):
        return 3
    elif ('No ' in cost) & ('Worse' in value):
        return 2
    elif ('Greater' in cost) & ('Worse' in value):
        return 1
    else:
        return None
```

```python
merged_hospital_df['Overall Score'] = merged_hospital_df[['Payment Category_x', 'Value Category']].apply(lambda x: get_overall_score(*x), axis=1)
```
